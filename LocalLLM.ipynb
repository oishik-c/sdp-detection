```python
import os
from dotenv import load_dotenv
from huggingface_hub import hf_hub_download

%load_ext autoreload
%autoreload 2

load_dotenv()
```

```python
HUGGINGFACE_TOKEN = os.environ.get("HUGGINGFACE_TOKEN")
```

```python
model_id = "rjmalagon/Nxcode-CQ-7B-orpo-Q8_0-GGUF"

filenames = [
    "nxcode-cq-7b-orpo.Q8_0.gguf",
]
```

```python
for filename in filenames:
    downloaded_file_path = hf_hub_download(
        repo_id=model_id,
        filename=filename,
        token=HUGGINGFACE_TOKEN,
        cache_dir="./models/"
    )
    print(downloaded_file_path)
```

```python
from llama_cpp import Llama
# n_threads = os.cpu_count() // 4

llm = Llama(
    model_path="models/models--TheBloke--deepseek-coder-6.7B-instruct-GGUF/snapshots/9e221e6b41cb1bf1c5d8f9718e81e3dc781f7557/deepseek-coder-6.7b-instruct.Q8_0.gguf",
    seed=1337,
    chat_format="chatml",  # ChatML format for this model
    n_ctx=4096,
    # n_threads=n_threads,
    n_batch=1024,
    n_ubatch=512,
    verbose=True,
)
```

```python
type(llm)
```

```python
output = llm.create_chat_completion(
    messages=[
        {
            "role": "system",
            "content": "You're an AI coding assistant who helps in solving coding questions",
        },
        {"role": "user", "content": "Write a code to calculate factorial of an input number in python"},
    ],
    temperature=0.3,
    max_tokens=512,
    top_p=0.95,
    
)
```
